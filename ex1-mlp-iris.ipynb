```python
import numpy as np
from tensorflow import keras
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Ada
```

```python
Attribute Information:

sepal length in cm
sepal width in cm
petal length in cm
petal width in cm
class:

Iris Setosa
Iris Versicolour
Iris Virginica
```

```python
iris_data = load_iris() # load the iris dataset
x = iris_data.data
y = iris_data.target.reshape(-1, 1) # Convert data to a single column
```

```python
y = keras.utils.to_categorical(y)
y
```

```python
# Split the data for training and testing
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20)
```

# <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">ШіЩҲШ§Щ„ 2:</div>
<div style="direction:rtl;text-align:right;font-family:Tahoma">
ЫҢЪ© ШҙШЁЪ©ЩҮ ШЁШ§ ШҜЩҲ hidden-layer ШҜШұ ЩҮШұ Щ„Ш§ЫҢЩҮ 10 ЩҶЩҲШұШ§ЩҶ ЩҲ ШӘШ§ШЁШ№ ЩҒШ№Ш§Щ„ЫҢШӘ relu ШЁШіШ§ШІЫҢШҜ. ЫҢЪ© Щ„Ш§ЫҢЩҮ Dropout ШЁШ§ ЩҶШұШ® 0.5 ШҜШұ Щ„Ш§ЫҢЩҮ ШўШ®Шұ Щ…Ш§ЩӮШЁЩ„ softmax ЩҶЫҢШІ Ш§Ш¶Ш§ЩҒЩҮ Ъ©ЩҶЫҢШҜ.
</div>

```python
# Build the model
model = Sequential()
model.add(Dense(10,activation='relu',input_dim=4))
model.add(Dense(10,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3,activation='softmax'))
```

```python
model.summary()
```

```python
# Adam optimizer with learning rate of 0.001
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(train_x,train_y,epochs=200,batch_size=5)
```

```python
# Test on unseen data
results = model.evaluate(test_x, test_y)

print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```
